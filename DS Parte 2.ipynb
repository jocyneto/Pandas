```python
def km_media(dataset, ano_atual):
    lista = []
    for item in dataset.items():
        result = item[1]['km'] / (ano_atual - item[-1]['ano'])
        lista.append(result)
    return lista
```

```python
valores = km_media(dados, 2021)
valores
```

```python
dados = {
    'Crossfox': {'km': 35000, 'ano': 2005}, 
    'DS5': {'km': 17000, 'ano': 2015}, 
    'Fusca': {'km': 130000, 'ano': 1979}, 
    'Jetta': {'km': 56000, 'ano': 2011}, 
    'Passat': {'km': 62000, 'ano': 1999}
}
```

```python
def km_media(dataset, ano_atual):
    dicionario = {}
    for item in dataset.items():
        result = item[1]['km'] / (ano_atual - item[-1]['ano'])
        item[1].update({'km_media': result})
        dicionario.update({ item[0]: item[1] })
    return dicionario
```

```python
km_media(dados, 2021)
```
